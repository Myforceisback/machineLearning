```python
from ucimlrepo import fetch_ucirepo
import pandas as pd

horse_colic = fetch_ucirepo(id=47)

X = horse_colic.data.features
y = horse_colic.data.targets

data_horse_colic = pd.concat([X, y], axis=1)

data_horse_colic
```

```python
target_column = y.columns[0]

grouped_by_result = data_horse_colic.groupby(target_column)

for name, group in grouped_by_result:
    print(f"\nРезультат: {name}")
    print(f"Количество строк: {len(group)}")
```

```python
result_groups = data_horse_colic.groupby(target_column)

first_result_name = list(result_groups.groups.keys())[0]

first_group = result_groups.get_group(first_result_name)

first_value = first_group.iloc[0, 0]

filtered_group = first_group[first_group.iloc[:, 0] == first_value]

print(f"Группа: {first_result_name}")
print(f"Значение первого столбца: {first_value}")
print(f"Количество строк: {len(filtered_group)}")

print(filtered_group.head())
```
