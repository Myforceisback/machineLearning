```python
import pandas as pd

df = pd.read_csv('/home/egor/machineLearing/lab7/tic_tac_toe_cleaned.csv', delimiter=',')
df_reduced = pd.read_csv(
    "/home/egor/machineLearing/lab7/tic_tac_toe_reduced.csv",
    delimiter=","
)

print(df.shape)
```

```python
categorical_columns = df.select_dtypes(include=['object']).columns
numeric_columns = df.select_dtypes(include=['number']).columns

target_column = 'class'

print("Категориальные данные:", categorical_columns)
print("Числовые данные:", numeric_columns)
print("Целевая переменная:", target_column)
print("Размер датасета:", df.shape)
print("Пропуски по столбцам:")
print(df.isnull().sum())
```

```python
X = df.drop('class', axis=1)
y = df['class'].map({
    'negative': 0,
    'positive': 1
})
X_reduced = df_reduced.drop("class", axis=1)

y_reduced = df_reduced["class"].map({
    "negative": 0,
    "positive": 1
})
X_encoded = pd.get_dummies(X, dtype=int)
X_reduced_encoded = pd.get_dummies(X_reduced, dtype=int)

df_reduced_encoded = pd.concat(
    [X_reduced_encoded, y_reduced.rename("class")],
    axis=1
)
df_encoded = pd.concat([X_encoded, y.rename('class')], axis=1)
df_encoded.head()
print(df_encoded.shape)

```
