```python
import pandas as pd
from ucimlrepo import fetch_ucirepo

horse_colic = fetch_ucirepo(id=47)

X = horse_colic.data.features.copy()
y = horse_colic.data.targets.copy()

df = pd.concat([X, y], axis=1)

df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace('-', '_', regex=False)
    .str.replace(' ', '_', regex=False)
)

target_column = y.columns[0]
target_column = target_column.strip().lower().replace('-', '_').replace(' ', '_')

df_reduced = pd.read_csv("horse_colic_reduced.csv")

print(df.shape)
print(df.head())
```

```python
categorical_columns = df.select_dtypes(include=['object']).columns
numeric_columns = df.select_dtypes(include=['number']).columns

print("Категориальные данные:", categorical_columns)
print("Числовые данные:", numeric_columns)
print("Целевая переменная:", target_column)
print("Размер датасета:", df.shape)
print("Пропуски по столбцам:")
print(df.isnull().sum())
```

```python
X = df.drop(target_column, axis=1)
y = df[target_column]

X_reduced = df_reduced.drop(target_column, axis=1)
y_reduced = df_reduced[target_column]

X_encoded = pd.get_dummies(X, dtype=int)
X_reduced_encoded = pd.get_dummies(X_reduced, dtype=int)

df_encoded = pd.concat(
    [X_encoded, y.rename(target_column)],
    axis=1
)

df_reduced_encoded = pd.concat(
    [X_reduced_encoded, y_reduced.rename(target_column)],
    axis=1
)

print(df_encoded.shape)
df_encoded.head()
```

```python
df_encoded.to_csv("horse_colic_normalized.csv", index=False)

df_reduced_encoded.to_csv(
    "horse_colic_reduced_encoded.csv",
    index=False
)

print(df_reduced_encoded.shape)
df_reduced_encoded.head()
```
